## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""

p = 60/100
q = 40/100

print(f"The probability that the fruit is an apple is {p} and the probability that the fruit is an orange is {q}.")
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# What is the probability that the first 5 fruits are all apples?
p_five = 0.6**5

print(f"The probability that the first 5 fruits are apples is {p_five}.")

# What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?
five_apples_fifteen_oranges = p_five*((0.4)**15)

print(f"The probability that the first 5 fruits are apples and the next 15 fruits are all oranges is {five_apples_fifteen_oranges}.")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# 20 samples of 100

# probability to get 5 apples and 15 oranges:

from scipy.special import comb

# Define the parameters
n = 20  # Total number of draws
k = 5   # Number of apples
p = 0.6 # Probability of drawing an apple 

# Calculate the probability using the binomial formula
probability_five_apples_fifteen_oranges = comb(n, k) * (p ** k) * ((1 - p) ** (n - k))

print(f"Probability of drawing exactly 5 apples and 15 oranges: {probability_five_apples_fifteen_oranges:.6f}")
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# Formula:
# P(X < 5) = \sum_{i=0}^{4} \binom{n}{i} p^i (1-p)^{n-i}

from scipy.stats import binom

n = 20   # Total number of draws
p = 0.6  # Probability of drawing an apple
k = 4    # We want the probability of getting less than 5 apples (0 to 4 apples)

# Calculate the cumulative probability
probability_less_than_5 = binom.cdf(k, n, p)

# Display the result
print(f"Probability of drawing fewer than 5 apples: {probability_less_than_5:.6f}")
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

# Generate x values (possible number of successes)
x = np.arange(0, n+1)

# Calculate the probability mass function (PMF) for each x value
pmf_values = binom.pmf(x, n, p)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.title('Probability Distribution Function (PDF) for Binomial Distribution')
plt.xlabel('Number of Apples Picked')
plt.ylabel('Probability')
plt.xticks(x)  # Ensure all x values are shown
plt.grid(axis='y', linestyle='--')
plt.show()

````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
import math

# Given parameters
lambda_ = 2.3  # Average number of goals per match
k = 5          # Number of goals we are interested in

# Applying the Poisson formula
probability = (lambda_**k * math.exp(-lambda_)) / math.factorial(k)

# Display the result
print(f"Probability of scoring exactly {k} goals: {probability:.6f}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
# Please label the axes and give a title to the plot 

import matplotlib.pyplot as plt
from scipy.stats import poisson

# Given parameters
lambda_ = 2.3  # Average number of goals per match
k_values = np.arange(0, 11)  # Number of goals from 0 to 10

# Calculate the Poisson PMF for each k value
pmf_values = poisson.pmf(k_values, lambda_)

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.xlabel('Number of Goals')
plt.ylabel('Probability')
plt.title('Poisson Probability Distribution ( = 2.3)')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--')
plt.show()
````
